```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
haber=pd.read_csv("haberman.csv")
print(haber.shape)
```

```python
print(haber.head(10))
```

```python
haber["Surv_status"].value_counts()
```

```python
sns.pairplot(haber,hue="Surv_status",height=3)
```

```python
a=np.mean(haber[:])
b=np.std(haber[:])
print(a)
print(b)
```

```python
haber.describe()
```

```python
#Histograms and Probability Density Function
sns.FacetGrid(haber,hue="Surv_status",height=5).map(sns.distplot,"age").add_legend()
plt.show()
```

```python
sns.FacetGrid(haber,hue="Surv_status",height=5).map(sns.distplot,"axil_nodes").add_legend()
plt.show()
```

```python
counts,bin_edges = np.histogram(haber['age'],bins=10,density=True)

pdf = counts /sum(counts)
print(pdf)
print(bin_edges)

cdf = np.cumsum(pdf)#cumulative sum
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
```
