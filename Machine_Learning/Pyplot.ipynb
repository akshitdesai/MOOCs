```python
import numpy as np
t=np.arange(0.,5.,0.2)

plt.plot(t,t**2,'r--',label="^2")
plt.plot(t,t**2.2,'bs',label='^2.2')
plt.plot(t,t**2.5,'y^',label='^2.5')
plt.grid()
plt.legend()  #add legand based on lines labels
plt.show()
```

```python
x=[1,2,3,4]
y=[1,4,9,16]
plt.plot(x,y,linewidth=5.0)
plt.show()
```

```python
x1=[1,2,3,4]
y1=[1,4,9,16]
x2=[1,2,3,4]
y2=[2,4,6,8]
lines=plt.plot(x1,y1,x2,y2)
plt.setp(lines[0],color='r',linewidth=2.0)
#MATLAB style arguments
plt.setp(lines[1],"color",'g','linewidth',2.0)
plt.grid()
```

```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)

t1=np.arange(0.,5.,0.1)
t2=np.arange(0.,5.,0.02)

plt.figure(1)


plt.subplot(211)
plt.grid()
plt.plot(t1,f(t1),'b-')

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2),"r--")
plt.show()
```
