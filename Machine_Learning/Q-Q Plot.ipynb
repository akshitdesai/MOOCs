```python
#Q-Q plot
import numpy as np
import pylab
import scipy.stats as stats

std_normal = np.random.normal(loc=0,scale=1,size=1000)
#loc=mean______scale=std-dev__________size=range

for i in range(0,101):
    print(i,np.percentile(std_normal,i))
    
```

```python
measurements = np.random.normal(loc=20,scale=5,size=50000)
#size should be more 


stats.probplot(measurements,dist="norm",plot=pylab)
pylab.show()
```

```python
#generate 100 sample form N(10,5)
measurnments = np.random.uniform(low=-1,high=1,size=100)

stats.probplot(measurnments,dist="norm",plot=pylab)
pylab.show()

#x-Gaussian disb.
#y-Uniform disb.
```
